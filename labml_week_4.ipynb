```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel('Data_lab_w4.xlsx')
df
```

```python
x = np.array(df['X']).reshape((-1, 1))
y = np.array(df['Y'])
model = LinearRegression().fit(x, y)
r_sq = model.score(x, y)
print(str(x.mean()) + '\n' + str(y.mean()) + '\n' + str(model.coef_) + '\n' + str(model.intercept_) + '\n' + str(r_sq))
```

```python
DATA = pd.read_csv("candy-data.csv", delimiter=',', index_col='competitorname')
DATA
```

```python
train_data = DATA.drop(['Caramel Apple Pops','Nestle Crunch'])
train_data
```

```python
X = pd.DataFrame(train_data.drop(['winpercent', 'Y'], axis=1))
y = pd.DataFrame(train_data['winpercent'])
```

```python
lr = LinearRegression().fit(X, y)
```

```python
CarAppPops = DATA.loc['Caramel Apple Pops',:].to_frame().T

lr.predict(CarAppPops.drop(['winpercent', 'Y'], axis=1))
```

```python
NestleCrunch = DATA.loc['Nestle Crunch',:].to_frame().T

lr.predict(NestleCrunch.drop(['winpercent', 'Y'], axis=1))
```
